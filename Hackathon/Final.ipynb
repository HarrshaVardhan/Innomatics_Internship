```python
# Defining a function for cleaning the 'Wage' & 'Value' columns

def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)

# applying the function to the wage column

data['Value'] = data['Value'].apply(lambda x: extract_value_from(x))
data['Wage'] = data['Wage'].apply(lambda x: extract_value_from(x))

data['Wage'].head()
```

```python
def convert_weights(s):
    s=s.replace("lbs","")
    return int(s)
```

```python
data['Weight'] = data['Weight'].apply(lambda x: convert_weights(x))
```

```python
def convert_heights(s):
    l = s.split("'")
    x = l[0]
    y = l[1]
    return int(x)*12 + int(y)
```

```python
data['Height'] = data['Height'].apply(lambda x: convert_heights(x))
```

```python
# Importing the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data.head()
```

```python
data = data[data.Overall > 86]

data = data.fillna(data.mean())
```

```python
# Collecting for further purpose

names = data.Name.tolist()
```

```python
# Dropping the categorical columns with in the data i.e Using numerical columns to create clusters

data = data.drop(['Name','ID', 'Nationality', 'Club', 'Preferred Foot', 'Work Rate',
       'Position', 'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM',
       'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM',
       'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB'], axis=1)
```

```python
# Glance of data after dropping the columns

data.head()
```

```python
# Standardizing the data
from sklearn import preprocessing
# Numpy array
x = data.values 
# Scaling the data
scaler = preprocessing.MinMaxScaler()
# Transforming the data
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
# Using PCA to reduce the columns into 2 columns

from sklearn.decomposition import PCA
pca = PCA(n_components = 2) 
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
# Applying the DBSCAN

from sklearn.cluster import DBSCAN

# Train the model using DBSCAN
db = DBSCAN(eps=1, min_samples=5)

# Prediction for dbscan clusters
db_clusters = db.fit_predict(reduced)
```

```python
# Making a dataframe by adding the names of plaayers with the clusters

reduced['cluster'] = db_clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name']
reduced.head()
```

```python
# Plotting DBSCAN

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced, legend=False,
                   fit_reg=False, size = 10, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(reduced.x, reduced.y, reduced.name):
    texts.append(plt.text(x, y, s))

ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel("x", fontsize = 20)
plt.ylabel("y", fontsize = 20)

plt.show()
```

```python
# Find an Optimal Illusion

from sklearn.neighbors import NearestNeighbors

# Calculate the distance from each point to its closest neighbor
nn = NearestNeighbors(n_neighbors = 2)

# Fit the nearest neighbor
reduced = pd.DataFrame(pca.fit_transform(X_norm))
nbrs = nn.fit(reduced)

# Returns two arrays - distance to the closest n_neighbors points and index for each point
distances, indices = nbrs.kneighbors(reduced)

# Sort the distance and plot it
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
plt.show()
```

```python
# Applying the DBSCAN using New Eps

from sklearn.cluster import DBSCAN

# Train the model using DBSCAN
db= DBSCAN(eps=0.3, min_samples=5)

# Prediction for dbscan clusters
db_clusters = db.fit_predict(reduced)

reduced['cluster'] = db_clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name']
reduced.head()
```

```python
# Plotting with Eps = 0.3

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced, legend=False,
                   fit_reg=False, size = 9, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(reduced.x, reduced.y, reduced.name):
    texts.append(plt.text(x, y, s))

ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=10)
plt.xlabel("x1", fontsize = 20)
plt.ylabel("y1", fontsize = 20)

plt.show()
```

Steps Done :

- Used DBScan to cluster the data
- DBScan requires 2 parameters - epsilon and minPts. Done the experiments performed to arrive at the final epsilons and minPts.
- Created clusters and plotted the data.
